# Question 1 - How Popular is Tennis?

```python
Y = [atp2000,atp2001,atp2002,atp2003,atp2004,atp2005,atp2006,atp2007,atp2008,atp2009,atp2010,atp2011,atp2012,atp2013,atp2014,atp2015,atp2016,atp2017,atp2018,atp2019,atp2020,atp2021,atp2022,atp2023,atp2024]

Ystr = []
for i in range(25):
  if len(str(i))==1:
    Ystr.append("atp200" + str(i))
  else:
    Ystr.append("atp20" + str(i))

Location_Count  = []
for i in range(len(Y)):
  Location_Count.append(len(list(set(Y[i]['Location']))))

Tournament_Count = []
for i in range(len(Y)):
  Tournament_Count.append(len(list(set(Y[i]['Tournament']))))

fig = plt.figure(figsize = (22,8))
ax = fig.add_subplot(1,1,1)
ax.plot(Ystr,Location_Count,'b',linewidth=4,marker='o',label='No. of Locations')
ax.plot(Ystr,Tournament_Count,'r',linewidth=4,marker='o',label='No of Tournaments')
ax.set_xlabel("Year")
ax.set_ylabel("Number of Locations(b)/Tournaments(r)")
ax.set_title("Question 1")
ax.grid(True)
ax.legend()
```

# Question 2 - What has been the World Rank of a Wimbeldon winner over the years?

```python
#Wimbeldon is the most popular Tennis Tournament
#It would be interesting to see what was the international rank of a Wimbeldon Winner of years 2000 to 2024
```

```python
Y = [atp2000,atp2001,atp2002,atp2003,atp2004,atp2005,atp2006,atp2007,atp2008,atp2009,atp2010,atp2011,atp2012,atp2013,atp2014,atp2015,atp2016,atp2017,atp2018,atp2019,atp2020,atp2021,atp2022,atp2023,atp2024]
Ystr = []
for i in range(25):
  if len(str(i))==1:
    Ystr.append("atp200" + str(i))
  else:
    Ystr.append("atp20" + str(i))

```

```python
Winners = []
WRanks = []
Years = []
for i in range(len(Y)):
  #The Winner
  Winner = Y[i]['Winner'].where((Y[i]['Tournament']=='Wimbledon') & (Y[i]['Round']=='The Final'))
  Winner = list(filter(lambda x: not (str(x) == 'nan' or str(x) == 'NaN'), Winner))

  #The Rank of the Winner
  WRank_Wimbledon = Y[i]['WRank'].where((Y[i]['Tournament']=='Wimbledon') & (Y[i]['Round']=='The Final'))
  WRank_Wimbledon = list(filter(lambda x: not (str(x) == 'nan' or str(x) == 'NaN'), WRank_Wimbledon))

  #The Year
  Year = Ystr[i]

  #In some years, Wimbledon wasn't played so important to check
  if len(WRank_Wimbledon)==0:
    continue
  else:
    Winners.append(Winner[0])
    WRanks.append(WRank_Wimbledon[0])
    Years.append(Year)
```

```python
fig = plt.figure(figsize = (22,8))
ax = fig.add_subplot(1,1,1)
ax.plot(Years,WRanks,'b',linewidth=4,marker='o')
ax.set_xlabel("Year")
ax.set_ylabel("World Rank of Players")
ax.set_title("Question 2")
ax.grid(True)
```

Answer - The Above plot

Useful Information:

1. The prestigious nature of Wimbledon is seen in the world rank of its winners, hich is usually 1 or 2,i.e., Winners of this tournament are the best or second best in the world

2. During 2001 and 2008, unexpected spikes are seen in the data and the world rank of Wimbledon winners went down to 125 and 20 respectively

# Question 3 - Which are the top 3 Tennis players?

Here we judge the most successful players in tennis with the amount of tournaments they have won in the years between 2000 and 2024

```python
#Defining a function that gets count of each entity in a list where entities are repeated
def count_names(names_list):
    name_count = {}
    for name in names_list:
        if name in name_count:
            name_count[name] += 1
        else:
            name_count[name] = 1
    return name_count
```

```python
#Getting all winners
W = []
for i in range(len(Y)):
  Winners = Y[i]['Winner'].where(Y[i]['Round']=='The Final')
  Winners = list(filter(lambda x: not (str(x) == 'nan' or str(x) == 'NaN'), Winners))
  for i in Winners:
    W.append(i)
```

```python
#Getting the number of tournaments won
Winners = count_names(W)
Winners= dict(sorted(Winners.items(), key=lambda item: item[1],reverse=True))
print(Winners)
```

```python
#Plotting
x  = list(Winners.keys())
y = list(Winners.values())
fig = plt.figure(figsize = (22,8))
ax = fig.add_subplot(1,1,1)
ax.plot(x[:11],y[:11],'b',linewidth=4,marker='o')
ax.set_xlabel("Player")
ax.set_ylabel("Number of Tournaments Won")
ax.set_title("Question 3")
ax.grid(True)
ax.legend()

for i in range(len(y[:11])):
    plt.text(x[i], y[i], str(y[i]), ha='center', va='bottom',fontsize=15)
```

# Question 4 - Which Winner of the Wimbledon was most expected to lose?

# Question 8 - What is the average weight of the best Tennis Players?

Here, The Best players are the ones who have won the most number of tournaments. We shall take a sample of the top 20 Tennis players from the data available

```python
#Defining a function that gets count of each entity in a list where entities are repeated
def count_names(names_list):
    name_count = {}
    for name in names_list:
        if name in name_count:
            name_count[name] += 1
        else:
            name_count[name] = 1
    return name_count
```

```python
#Getting all winners
W = []
for i in range(len(Y)):
  Winners = Y[i]['Winner'].where(Y[i]['Round']=='The Final')
  Winners = list(filter(lambda x: not (str(x) == 'nan' or str(x) == 'NaN'), Winners))
  for i in Winners:
    W.append(i[:-1])
```

```python
#Getting the number of tournaments won
Winners = count_names(W)
Winners= dict(sorted(Winners.items(), key=lambda item: item[1],reverse=True))
Winners = list(Winners.keys())[:21]
print(Winners)
```

```python
TBD = []
Weight_Dict={}
for i in range(len(Winners)):
  weight = atpp['weight_kg'].where(atpp['full_name']==Winners[i])
  weight = list(filter(lambda x: not (str(x) == 'nan' or str(x) == 'NaN'), weight))
  if len(weight)==0:
    continue
  else:
    Weight_Dict[Winners[i]] = weight[0]

average_weight = np.mean(list(Weight_Dict.values()))
average_weight = np.round(average_weight,2)
print("The Average Weight of an average pro Tennis player is "  + str(average_weight) + " kilograms")
```
